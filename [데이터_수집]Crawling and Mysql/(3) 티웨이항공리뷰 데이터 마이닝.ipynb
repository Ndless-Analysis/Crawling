```python
import nltk
from konlpy.tag import Twitter
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/gulim.ttc").get_name()
rc('font',family = font_name)

from wordcloud import WordCloud
import matplotlib.pyplot as plt

t = Twitter()
```

```python
import nltk
from konlpy.tag import Twitter; t = Twitter()
```

## 2-1. 리뷰 제목 분석

```python
review_title = data['title']
review_title
```

```python
title_text = ''

for each_line in review_title:
    title_text = title_text + each_line +'\n'
```

```python
tokens_ko = t.morphs(title_text)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(100)
```

```python
stop_words = ['.','적','한','에','중','이','은','..','으로','이었다','는','...','도', '함','의','인','(',')','가','-','다','ㅜㅜ','의','~','ㅡ',',','!']
tokens_kor = [each_word for each_word in tokens_ko 
             if each_word not in stop_words]

ko = nltk.Text(tokens_kor)
ko.vocab().most_common(100)
```

```python
words = ko.vocab()
words
```

```python
import graphviz
from sklearn.tree import export_graphviz
plt.figure(figsize=(15,6))
ko.plot(50)
plt.show()
```

```python
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
                     relative_scaling=0.2,
                     background_color='white').generate_from_frequencies(words)

plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

# 전체 리뷰 제목 분석 결과,

대체적으로 가격 대비 만족스럽다는 의견, 그러나 '좌석 좁음', '지연' 등과 같은 단점을 상쇄하기 위한 마케팅 전략의 필요성이 보임
